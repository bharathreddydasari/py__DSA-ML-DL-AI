```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for some value to get same result every time 
numpy.random.seed(7)
```

```python
dataset = numpy.loadtxt(r"C:\Users\dasarred\Documents\pima-indians-diabetes.csv", delimiter=",")#
```

```python
dataset.size
```

```python
dataset.shape
```

```python
X = dataset[:,0:8]#input
Y = dataset[:,8]#output
```

```python
X.shape
```

```python
Y.shape
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
predictions = model.predict(X)
```

```python
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
